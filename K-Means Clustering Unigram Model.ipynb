```python
import json
import random
```

```python
data = []
ids = []
titles = []
with open("with_abstracts.json", 'r') as json_file:
    for line in json_file.readlines():
        if random.random() < 0.08:
            data.append(json.loads(line).get("abstract"))   
            ids.append(json.loads(line).get("recid"))  
            titles.append(json.loads(line).get("title"))
```

```python
from __future__ import print_function
from sklearn.datasets import fetch_20newsgroups
from sklearn.decomposition import TruncatedSVD
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import HashingVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import Normalizer
from sklearn import metrics 

from sklearn.cluster import KMeans, MiniBatchKMeans

import logging
from optparse import OptionParser
import sys
from time import time

import numpy as np
```

```python
vectorizer = TfidfVectorizer(max_df=0.95, max_features=100, min_df=2, stop_words='english')
```

```python
X=vectorizer.fit_transform(data)
```

```python
X[23,:].toarray()
```

```python
km = KMeans(n_clusters=100, init='k-means++', max_iter=100, n_init=1,verbose=True)
```

```python
km.fit(X)
```

```python
print("Clustering sparse data with %s" % km)
```

```python
print("Top terms per cluster:")
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()
for i in range(100):
    print("Cluster %d:" % i, end='')
    for ind in order_centroids[i, :10]:
        print(' %s' % terms[ind], end='')
    print()
    
```

```python

for i in range(len(data)):
    if km.labels_[i]==60:
        print(" ".join(["Cluster:", str(km.labels_[i])]))
        print(titles[i])
        print(" ".join(["RecID:", str(ids[i])]))
```
